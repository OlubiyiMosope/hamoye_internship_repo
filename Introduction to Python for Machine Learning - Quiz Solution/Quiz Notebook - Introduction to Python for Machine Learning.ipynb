### Question Solutions:

##### 11. What is the total sum of Animal Fat produced in 2014 and 2017 respectively?

```python
df.groupby('Item')[['Y2014','Y2017']].sum()
```

##### 12. What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

```python
print(f"mean: {(df['Y2015'].mean()):.3f}, std: {(df['Y2015'].std()):.3f}")
```

##### 13. What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?

```python
df['Y2015'].isnull().sum()
```

```python
round((df['Y2015'].isnull().sum() / df['Y2015'].shape[0] * 100), 2)
```

##### 15. What year has the highest sum of Import Quantity?

```python
element_df = df.groupby('Element')[['Y2014', 'Y2015','Y2016','Y2017','Y2018']].sum()
element_df
```

```python
# max(element_df[element_df.index == 'Import Quantity'])
max_col = np.argmax(element_df[element_df.index == 'Import Quantity'].values[0])
print(max_col)
```

```python
element_df[element_df.columns[max_col]][element_df.index == 'Import Quantity']
```

##### 16. What is the total number of the sum of Production in 2014?

```python
element_df['Y2014'][element_df.index == 'Production']
```

##### 17. Which of these elements had the highest sum in 2018?

```python
element_df[element_df['Y2018'] == max(element_df['Y2018'])]
```

##### 18. Which of these elements had the 3rd lowest sum in 2018?

```python
df.groupby('Element')[['Y2018','Element']].sum()
```

```python
(df.groupby('Element')[['Y2018','Element']].sum())['Y2018'].nsmallest(3)
```

##### 19. What is the total Import Quantity in Algeria in 2018?

```python
df.groupby(['Area','Element'])['Y2018'].sum().unstack()
```

```python
df.pivot_table('Y2018', index='Area', columns='Element', aggfunc='sum')
```

##### 20.  What is the total number of unique countries in the dataset?
